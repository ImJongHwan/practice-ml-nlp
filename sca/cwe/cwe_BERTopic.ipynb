```python
# 문장 토큰화

import nltk
import csv
from nltk.tokenize import sent_tokenize
from nltk.tokenize import TreebankWordTokenizer
from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups

nltk.download('punkt')
```

```python
csv_file = open("cwe_preprocessed.csv", "r")
f = csv.reader(csv_file, delimiter=",", doublequote=True, lineterminator="\r\n", quotechar='"', skipinitialspace=True)
```

```python
word_tokenizer = TreebankWordTokenizer()
contents = []
sentences = []
next(f)
for row in f:
  if len(row) > 3 :
    content = []
    content.append(row[0])
    content.append(row[1])
    content.append(word_tokenizer.tokenize(row[2]))
    expended_sen = []
    for sen in sent_tokenize(row[3]):
      expended_sen.append(word_tokenizer.tokenize(sen))
    content.append(expended_sen)
    sentences.append(row[3])
    contents.append(content)
```

```python
print(len(contents))
```

```python
model = BERTopic()
topics, probabilities = model.fit_transform(sentences)

print('각 문서의 토픽 번호 리스트: ', len(topics))
print('첫번째 문서의 토픽 번호: ', topics[0])
```
